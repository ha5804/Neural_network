```python
# data1, data2, data3, data4: shape (1000, 12, 12)
# flatten
X = np.concatenate([
    data1.reshape(1000, -1),
    data2.reshape(1000, -1),
    data3.reshape(1000, -1),
    data4.reshape(1000, -1)
], axis=0)  # shape: (4000, 144)

y = np.array([0]*1000 + [1]*1000 + [2]*1000 + [3]*1000).reshape(-1, 1)  # shape: (4000, 1)

model = MLP(input_size=144, hidden_size=64, num_label=4)

# epoch 학습
for epoch in range(100):
    loss = model.train(X, y)
    if (epoch + 1) % 10 == 0:
        pred = model.predict(X)
        acc = np.mean(pred.reshape(-1, 1) == y)
        print(f"Epoch {epoch+1}, Loss: {loss:.4f}, Accuracy: {acc:.4f}")

predictions = model.predict(X[:10])
print("예측 결과:", predictions)
print("실제 정답:", y[:10].reshape(-1))
```
